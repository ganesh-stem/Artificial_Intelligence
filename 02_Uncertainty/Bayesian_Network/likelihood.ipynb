Bayesian Network: Data structure that represents the dependencies among randon variables.

For example, if it's rainy today, then the flights might get delayed or ger cancelled.
> - It uses Directed Graph.<br>
> - Each node represents a random variable.<br>
> - Arrow from X to Y means X is parent of Y.<br>
> - Each node X has probability distribution P(X | Parents(X)).
