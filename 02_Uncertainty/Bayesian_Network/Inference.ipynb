Bayesian Network: Data structure that represents the dependencies among randon variables.

For example, if it's rainy today, then the flights might get delayed or ger cancelled.
> - It uses Directed Graph.<br>
> - Each node represents a random variable.<br>
> - Arrow from X to Y means X is parent of Y.<br>
> - Each node X has probability distribution P(X | Parents(X)).

Inference:
> - Query X: Variable X for which to compute distribution. For example, the probability that you miss the train.<br><br>
> - Evidence variable E: Observed variables for event e. For example, you observed that it's raining. You also observed that it's light rain or heavy rain.<br><br>

> Using those evidence, you want to know the probabily that your train is delayed.

```
from model import model
```

```
# Calculate predictions
# Given the evidence that train is delayed. Calculate the value of the other
# Random variables in the bayesian network.
predictions = model.predict_proba({
    "train": "delayed"
})
```

```
# Print predictions for each node
for node, prediction in zip(model.states, predictions):
    if isinstance(prediction, str):
        print(f"{node.name}: {prediction}")
    else:
        print(f"{node.name}")
        for value, probability in prediction.parameters[0].items():
            print(f"    {value}: {probability:.4f}")
```

```
# Given the evidence: Heavy rain.
predictions = model.predict_proba({
    "rain": "heavy"})

# Print predictions for each node
for node, prediction in zip(model.states, predictions):
    if isinstance(prediction, str):
        print(f"{node.name}: {prediction}")
    else:
        print(f"{node.name}")
        for value, probability in prediction.parameters[0].items():
            print(f"    {value}: {probability:.4f}")
```

```
# Given the evidences: Heavy rain and train is delayed.
predictions = model.predict_proba({
    "rain": "heavy",
    "train": "delayed"
})

# Print predictions for each node
for node, prediction in zip(model.states, predictions):
    if isinstance(prediction, str):
        print(f"{node.name}: {prediction}")
    else:
        print(f"{node.name}")
        for value, probability in prediction.parameters[0].items():
            print(f"    {value}: {probability:.4f}")
```
